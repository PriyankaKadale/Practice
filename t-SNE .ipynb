```python
# load data
filename = 'TSNE_data.csv'
dataframe = pd.read_csv(filename)
```

```python
dataframe # Tumordata set
```

```python
# Split-out validation dataset
array = dataframe.values
# separate array into input and output components
X = array[:,1:] #without diagnosis column
Y = array[:,0] #diagnosis array 
```

```python
#TSNE visualization
from bioinfokit.visuz import cluster
#It is better to keep component value as 2 its easy to visualize
data_tsne = TSNE(n_components=2, perplexity=30).fit_transform(X) #can change perplexity values and see the result by opening the stored png file
cluster.tsneplot(score=data_tsne) #png file is created and stored automatically
```

```python
data_tsne # Since n_component =2 hence the data is in 2D
```

```python
# get a list of categories
color_class = dataframe['diagnosis'].to_numpy()
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15, 1))

#Plot will be stored in the default directory
```
